```python
def Return_Race_Chart():
    TRL = pd.read_csv('/home/jovyan/Data/Race.csv')
    Cor_Sub_Race = []
    Cor_Race = []
    Cor_Count = []

    for x1, y1 in TRL.iterrows():
        cont = TRL.at[x1,'Count']
        if cont > 0:
            SubR = TRL.at[x1,'Sub-Race']
            if not isinstance(SubR, str):
                #continue
                SubR = TRL.at[x1,'Source']+"."+TRL.at[x1,'Race']

            Cor_Sub_Race.append(SubR)
            Cor_Race.append(TRL.at[x1,'Race'])
            Cor_Count.append(cont)


    myset = set(Cor_Race)
    Cor_Race_Uni = list(myset)
    Fin_Sub_Race = ["Race"]
    Fin_Race = [""]
    Fin_Count = [0]

    for RU in Cor_Race_Uni:
        Fin_Sub_Race.append(RU)
        Fin_Race.append("Race")
        Fin_Count.append(0)

    Cor = {'SubRace': Cor_Sub_Race,
           'Race': Cor_Race,       
           'Count': Cor_Count
          }

    for RU in Cor_Sub_Race:
        Fin_Sub_Race.append(RU)
    for RU in Cor_Race:
        Fin_Race.append(RU)
    for RU in Cor_Count:
        Fin_Count.append(RU)

    Tem = {'SubRace': Fin_Sub_Race,
           'Race': Fin_Race,       
           'Count': Fin_Count
          }

    df = pd.DataFrame(Tem)
    df = df.iloc[::-1]

    df2 = df.query("Race!='Race'")

    for x1, y1 in df2.iterrows():
        df3 = df.query("Race=='Race' and SubRace=='"+y1['Race']+"'")
        df4 = df.query("Race=='"+y1['Race']+"'")
        if len(df4)==1 and len(df3)==1:
            for d1, d2 in df3.iterrows():
                df = df.drop(d1)
            df5 = pd.DataFrame(TRL.query("Race=='"+y1['Race']+"' and `Sub-Race`=='"+y1['SubRace']+"'"))        
            if len(df5)>0:
                df.at[x1,'SubRace']= df5.iloc[0]['Tag2']
            df.at[x1,'Race']='Race'

    df = df.iloc[::-1]
    df.to_csv('/home/jovyan/Data/Race_Burst.csv', index=False) #<<<<<<<
    #-------------------------------------------------------------------
    TGL = pd.read_csv('/home/jovyan/Data/Race_Burst.csv')
    fig = px.treemap(TGL,
                      names='SubRace',
                      parents='Race',
                      values='Count'
                     )
    fig.update_traces(root_color="lightgrey")
    fig.update_layout(margin = dict(t=0, l=0, r=0, b=0))
    fig.update_yaxes(tickformat="80%")
    fig.update_layout(height=500)
    return setup_ui(fig)   
```

```python
def Return_RC_BubCha():
    Tab_Rac = pd.read_csv('/home/jovyan/Data/Race.csv')
    Tab_Cla = pd.read_csv('/home/jovyan/Data/Class_Level.csv')
    Tab_Cha = pd.read_csv('/home/jovyan/Data/Characters.csv')
    data = []

    #-------------------------------------------
    List_Race1  = []
    for x1, y1 in Tab_Rac.iterrows():
        if Tab_Rac.at[x1,'Count'] > 0:
            List_Race1.append(Tab_Rac.at[x1,'Tag2'])
    List_Class1 = []
    for x1, y1 in Tab_Cla.iterrows():#Sub-Class
        if Tab_Cla.at[x1,'Class']==Tab_Cla.at[x1,'Tag3']:
            if Tab_Cla.at[x1,'C_Count'] > 0:
                List_Class1.append(Tab_Cla.at[x1,'Tag2'])

    #-------------------------------------------

    List_Class2 = []

    for x1, y1 in Tab_Cla.iterrows():
        if Tab_Cla.at[x1,'C_Count'] > 0:
            List_Class2.append(Tab_Cla.at[x1,'Tag2'])

    for c in List_Class1:
        List_Class2.remove(c)

    #-------------------------------------------

    Fin_Rac = []
    Fin_Cla = []
    Fin_Cou = []
    Fin_Typ = []

    for y1 in List_Race1:
        Ro = []
        for y2 in List_Class1:
            Con=0
            Con=len(Tab_Cha.query("Race=='"+y1+"' and `Class Breakdown`.str.contains('"+y2+"')").index)
            Fin_Rac.append(y1)
            Fin_Cla.append(y2)
            Fin_Cou.append(Con)
            Fin_Typ.append("Class")
            Ro.append(Con)

    for y1 in List_Race1:
        Ro = []
        for y2 in List_Class2:
            Con=0
            Con=len(Tab_Cha.query("Race=='"+y1+"' and `Class Breakdown`.str.contains('"+y2+"')").index)
            Fin_Rac.append(y1)
            Fin_Cla.append(y2)
            Fin_Cou.append(Con)
            Fin_Typ.append("SubClass")
            Ro.append(Con)

    Fin_SCla = []
    Fin_SSub = []

    for y1 in Fin_Cla:
        Ro=Tab_Cla.query("Tag2=='"+y1+"'").to_dict('list')
        Fin_SCla.append(Ro['Class'][0])
        Fin_SSub.append(Ro['Sub-Class'][0])

    Fin_Dict = {
      "Race": Fin_Rac,
      "Class": Fin_Cla,    
      "Count": Fin_Cou,
      "Type": Fin_Typ,
      "CClass": Fin_SCla,
      "CSubClass": Fin_SSub,
    }

    fdf = pd.DataFrame(Fin_Dict)
    for d1, d2 in fdf.iterrows():
        if d2['Count'] == 0:
            fdf = fdf.drop(d1)
    fdf = fdf.sort_values(['CClass', 'CSubClass'],ascending = [False, False])
    fdf = fdf.iloc[::-1]

    fig = px.scatter(fdf, x="Class", y="Race", size="Count",
                     color="CClass", color_continuous_scale="RdBu",
                     facet_col="Type")
    fig.update_layout(height=1000)
    fig.update_xaxes(range=[-1,len(List_Class1)-1], row=1, col=1)
    fig.update_yaxes(range=[-1,len([*set(Fin_Rac)])], row=1)
    fig.update_xaxes(matches=None)
    return setup_ui(fig)
```

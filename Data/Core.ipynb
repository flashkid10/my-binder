```python
import pandas as pd    
```

```python
Table_Race = []
def Process_Race():
    SHT_Characters = pd.read_csv('/home/jovyan/Data/Characters.csv')
    Table_Race = pd.read_csv('/home/jovyan/Data/Race.csv')
    
    for x2, y2 in Table_Race.iterrows():
        Table_Race.at[x2,'Count'] = 0

    for x1, y1 in SHT_Characters.iterrows():
        for x2, y2 in Table_Race.iterrows():
            #print(x2)
            if y2[4]==y1[5]:
                #y2[6]+=1
                Table_Race.at[x2,'Count'] += 1

    Table_Race.to_csv('/home/jovyan/Data/Race.csv', index=False) #<<<<<<<
```

```python
Table_Class_Level = []
def Process_Class_Level():
    SHT_Characters = pd.read_csv('/home/jovyan/Data/Characters.csv')
    Table_Class_Level = pd.read_csv('/home/jovyan/Data/Class_Level.csv')
    
    
    for x2, y2 in Table_Class_Level.iterrows():
        Table_Race.at[x2,'C_Count'] = 0
        Table_Race.at[x2,'L_Count'] = 0
    
    for x1, y1 in SHT_Characters.iterrows():
        for x2, y2 in Table_Class_Level.iterrows():
            if y2[4] in y1[6]:
                Table_Class_Level.at[x2,'C_Count'] += 1
    
    Table_Class_Level.to_csv('/home/jovyan/Data/Class_Level.csv', index=False) #<<<<<<<

Process_Class_Level()
```

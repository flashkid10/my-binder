```python

def Check_Race(SF, Nam):
    Counter=0;
    for x,y in SF:
        if y[1]==Nam:
            if y[2]>0:
                Counter+=1
    return Counter == 1
    


def Display_Race(Arr):
    SF = Table_Race[[Arr[0], Arr[1], Arr[2]]]
    
    
    SFL = []
    
    for x,y in SF.iterrows():
        if y[2] > 0:
            SFL.append(y)
            
    
    mat1 = np.array(SFL)

            
    print(mat1[:, 0])
    SF["All"] = "All" # in order to have a single root node
    
    
    
    fig = px.treemap(SF, path=['All', Arr[0], Arr[1]], values=Arr[2])
    fig.update_traces(root_color="lightgrey")
    fig.update_layout(margin = dict(t=0, l=0, r=0, b=0))
    fig.update_layout(autosize=False,width=1100,height=800)
    #fig.update_layout(autosize=True)
    return fig
```

```python
import pandas as pd
import plotly.express as px

SHT_Characters = pd.read_csv('/home/jovyan/Data/Characters.csv')
SHT_Race = pd.read_csv('/home/jovyan/Data/Race.csv')



#print(SHT_Race)

for x1, y1 in SHT_Characters.iterrows():
    for x2, y2 in SHT_Race.iterrows():
        #print(x2)
        if y2[4]==y1[5]:
            #y2[6]+=1
            SHT_Race.at[x2,'Count'] += 1

#All_Child = []
#All_Parents = []
            
            
#for x2, y2 in SHT_Race.iterrows():
    #SHT_Race.at[x2,'Count'] += 1

SF = SHT_Race[["Source", "Race", "Count"]]
SF["all"] = "all" # in order to have a single root node
fig = px.treemap(SF, path=['all', 'Source', 'Race'], values='Count')
fig.update_traces(root_color="lightgrey")

#fig.update_layout(margin = dict(t=-10, l=-10, r=-10, b=-10))
fig.show()
    
```

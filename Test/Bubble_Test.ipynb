```python
Tab_Rac = pd.read_csv('/home/jovyan/Data/Race.csv')
Tab_Cla = pd.read_csv('/home/jovyan/Data/Class_Level.csv')
Tab_Cha = pd.read_csv('/home/jovyan/Data/Characters.csv')

data = []


#-------------------------------------------
List_Race1  = []
for x1, y1 in Tab_Rac.iterrows():
    if Tab_Rac.at[x1,'Count'] > 0:
        List_Race1.append(Tab_Rac.at[x1,'Tag2'])
List_Class1 = []
for x1, y1 in Tab_Cla.iterrows():#Sub-Class
    if Tab_Cla.at[x1,'Class']==Tab_Cla.at[x1,'Tag3']:
        if Tab_Cla.at[x1,'C_Count'] > 0:
            List_Class1.append(Tab_Cla.at[x1,'Tag2'])
            
#-------------------------------------------
            


List_Class2 = []
for x1, y1 in Tab_Cla.iterrows():
    if Tab_Cla.at[x1,'C_Count'] > 0:
        List_Class2.append(Tab_Cla.at[x1,'Tag2'])

#-------------------------------------------


Fin_Rac = []
Fin_Cla = []
Fin_Cou = []
Fin_Typ = []






for y1 in List_Race1:
    Ro = []
    for y2 in List_Class1:
        Con=0
        Con=len(Tab_Cha.query("Race=='"+y1+"' and `Class Breakdown`.str.contains('"+y2+"')").index)
        Fin_Rac.append(y1)
        Fin_Cla.append(y2)
        Fin_Cou.append(Con)
        Fin_Typ.append("Class")
        Ro.append(Con)

for y1 in List_Race1:
    Ro = []
    for y2 in List_Class2:
        Con=0
        Con=len(Tab_Cha.query("Race=='"+y1+"' and `Class Breakdown`.str.contains('"+y2+"')").index)
        Fin_Rac.append(y1)
        Fin_Cla.append(y2)
        Fin_Cou.append(Con)
        Fin_Typ.append("SubClass")
        Ro.append(Con)



Fin_Dict = {
  "Race": Fin_Rac,
  "Class": Fin_Cla,
  "Count": Fin_Cou,
  "Type": Fin_Typ
}

fdf = pd.DataFrame(Fin_Dict)
```

```python
for d1, d2 in fdf.iterrows():
    if d2['Count'] == 0:
        fdf = fdf.drop(d1)


fig = px.scatter(fdf, x="Class", y="Race", size="Count",
                 color="Count", color_continuous_scale="Blues",
                 facet_col="Type", hover_name="Class")
#, log_x=True, size_max=60
fig.update_layout(height=1000)
fig.show()
```
